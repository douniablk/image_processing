```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r'C:\Users\DELL\Desktop\tp_tim\pics\bananapic.jpg', cv2.IMREAD_GRAYSCALE)

kernelx = np.array([[ -1, 0, 1],
                    [ -1, 0, 1],
                    [ -1, 0, 1]])

kernely = np.array([[ 1,  1,  1],
                    [ 0,  0,  0],
                    [-1, -1, -1]])

img_prewittx = np.zeros_like(img, dtype=np.float32)
img_prewitty = np.zeros_like(img, dtype=np.float32)

for i in range(1, img.shape[0]-1):
    for j in range(1, img.shape[1]-1):
        region = img[i-1:i+2, j-1:j+2]
        gx = np.sum(kernelx * region)
        gy = np.sum(kernely * region)
        img_prewittx[i, j] = gx
        img_prewitty[i, j] = gy


img_prewitt = np.sqrt(img_prewittx**2 + img_prewitty**2)



seuil = 100  
img_binaire = np.where(img_prewitt > seuil, 255, 0).astype(np.uint8)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1), plt.title('Original'), plt.imshow(img, cmap='gray')

plt.subplot(1,3,2), plt.title('countour perwitt'), plt.imshow(img_binaire, cmap='gray')
plt.show()
```
